# Analytical Questions

#### 1.  Average Monthly Temperature: What is the average temperature for each month?

```python
import pandas as pd
from datetime import datetime

# Define a function to parse dates with multiple formats
def parse_date(date_str):
    for fmt in ("%d-%m-%Y", "%m/%d/%Y"):
        try:
            return datetime.strptime(date_str, fmt)
        except ValueError:
            continue
    return None  # If no format matches, return None

#Load the data Set...
data = pd.read_csv("project 3 data.csv")

# Apply the parse_date function to the 'Day' column
data['Day'] = data['Day'].apply(parse_date)

# Drop rows where 'Day' could not be parsed
data = data.dropna(subset=['Day'])

# Extract the month and year from the 'Day' column
data['Month'] = data['Day'].dt.to_period('M')

# Calculate the average temperature for each month
average_monthly_temperature = data.groupby('Month')['Average temperature (°F)'].mean()

print("Average Monthly Temperature:")
print(average_monthly_temperature)
```

#### 5. Rainfall Patterns: How does monthly rainfall compare to the annual total?

```python
import pandas as pd
from datetime import datetime

# Define a function to parse dates with multiple formats
def parse_date(date_str):
    for fmt in ("%d-%m-%Y", "%m/%d/%Y"):
        try:
            return datetime.strptime(date_str, fmt)
        except ValueError:
            continue
    return None  # If no format matches, return None

# Load the CSV file
data = pd.read_csv("project 3 data.csv")

# Apply the parse_date function to the 'Day' column
data['Day'] = data['Day'].apply(parse_date)

# Drop rows where 'Day' could not be parsed
data = data.dropna(subset=['Day'])

# Extract the year and month from the 'Day' column
data['Year'] = data['Day'].dt.year
data['Month'] = data['Day'].dt.month

# Calculate the total rainfall for each month
monthly_rainfall = data.groupby(['Year', 'Month'])['Rainfall for month (in)'].sum().reset_index()

# Calculate the total rainfall for each year
annual_rainfall = data.groupby('Year')['Rainfall for year (in)'].max().reset_index()

# Merge monthly and annual rainfall data
rainfall_comparison = pd.merge(monthly_rainfall, annual_rainfall, on='Year', how='left')

# Calculate the percentage of annual rainfall that occurs each month
rainfall_comparison['Monthly vs Annual (%)'] = (rainfall_comparison['Rainfall for month (in)'] /
                                                rainfall_comparison['Rainfall for year (in)']) * 100

print("Rainfall Comparison:")
print(rainfall_comparison)
```

#### 10. Correlation Matrix: Create a heatmap to visualize the correlation matrix of the weather metrics.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
data = pd.read_csv("project 3 data.csv")

# Define a function to parse dates with mixed formats
def parse_date(date_str):
    for fmt in ('%d-%m-%Y', '%m/%d/%Y'):
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            continue
    return pd.NaT

# Apply the parsing function to the 'Day' column
data['Day'] = data['Day'].apply(parse_date)

# Check for any NaT values that couldn't be parsed
invalid_dates = data['Day'].isna().sum()
print(f"Number of invalid dates: {invalid_dates}")

# Drop rows with invalid dates if any
data = data.dropna(subset=['Day'])

# Compute the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=.5)

# Add title
plt.title('Correlation Matrix of Weather Metrics')

# Show the plot
plt.tight_layout()
plt.show()
```
